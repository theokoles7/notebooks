### A. Generation

```python
"""Generate datasets."""

from sklearn.datasets   import make_moons

# Generate datasets.
X_1, Y_1 =  make_moons(n_samples = 250, noise = 0.20, random_state = 7)
X_2, Y_2 =  make_moons(n_samples = 250, noise = 0.35, random_state = 7)

# Print shapes for verification.
print(f"Dataset 1 Samples:  {X_1.shape}")
print(f"Dataset 1 Targets:  {Y_1.shape}")
print(f"Dataset 2 Samples:  {X_2.shape}")
print(f"Dataset 2 Targets:  {Y_2.shape}")
```

### B. Plot

```python
"""Plot datasets."""

from matplotlib.pyplot  import figure, grid, scatter, show, subplot, title, xlabel, ylabel

# Initialize figure dimensions.
figure(figsize = (12, 5))

# Set subplot 1.
subplot(1, 2, 1)

# Create scatter plot of dataset 1, with noise = 0.20.
scatter(X_1[:, 0][Y_1==1], X_1[:, 1][Y_1==1], marker = "+", c = "purple")
scatter(X_1[:, 0][Y_1==0], X_1[:, 1][Y_1==0], marker = "o", c = "green")

# X contains two features, x1 and x2
xlabel(r"$x_1$", fontsize=20)
ylabel(r"$x_2$", fontsize=20)

# Display grid.
grid(visible = True)

# Set title.
title("Noise = 0.20")

# Set subplot 2.
subplot(1, 2, 2)

# Create scatter plot of dataset 2, with noise = 0.35.
scatter(X_2[:, 0][Y_2==1], X_2[:, 1][Y_2==1], marker = "+", c = "purple")
scatter(X_2[:, 0][Y_2==0], X_2[:, 1][Y_2==0], marker = "o", c = "green")

# X contains two features, x1 and x2
xlabel(r"$x_1$", fontsize=20)
ylabel(r"$x_2$", fontsize=20)

# Display grid.
grid(visible = True)

# Set title.
title("Noise = 0.35")

# Displaying the plot.
show()
```

### 1-C. Split

```python
"""Create training and test splits."""

from sklearn.model_selection    import train_test_split

# Split data into train and test sets for dataset 1.
x_1_train, x_1_test, y_1_train, y_1_test =  train_test_split(X_1, Y_1, test_size = 0.25)

# Split data into train and test sets for dataset 2.
x_2_train, x_2_test, y_2_train, y_2_test =  train_test_split(X_2, Y_2, test_size = 0.25)
```

## 2. KNN Classifier
* [Implementation](#2-a-implementation)
* [Training](#2-b-training)
* [Testing](#2-c-testing)

### 2-A. Implementation

```python
"""Implement KNN classifier."""

from sklearn.neighbors import KNeighborsClassifier

# Initialize KNN classifier.
knn_classifier: KNeighborsClassifier =  KNeighborsClassifier(n_neighbors = 1)
```

### 2-B. Training

```python
"""Train KNN classifier."""

# Fit the data to the classifier.
knn_classifier.fit(X = x_1_train, y = y_1_train)
```

### 2-C. Testing

```python
"""Test KNN classifier."""

from numpy              import ndarray
from sklearn.metrics    import accuracy_score

# Make predictions.
predictions:    ndarray =   knn_classifier.predict(X = x_1_test)

# Report accuracy.
print(f"KNN Classifier Accuracy: {accuracy_score(y_true = y_1_test, y_pred = predictions):.2f}")
```
